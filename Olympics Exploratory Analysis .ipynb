```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("athlete_events.csv")
df2 = pd.read_csv("noc_regions.csv")
```

```python
df.head()
```

```python
df2.head()
```

```python
merge_df= df.merge(df2, how = "left", on = "NOC")
merge_df.head()
```

```python
merge_df.shape
```

```python
merge_df.rename(columns={"region": "Region", "notes": "Notes"}, inplace=True)
```

```python
top_10_countries = merge_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=0)
plt.title("Overall Participation by country")
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette="husl");
```

```python
plt.figure(figsize=(15,10))
plt.title("Age distribution of the athletes")
plt.xlabel("Age")
plt.ylabel("No of Participants")
plt.hist(merge_df.Age, bins = np.arange(10,90,5), color="teal", edgecolor="black");
```

```python
winter_sports= merge_df[merge_df.Season == "Winter"].Sport.unique()
winter_sports
```

```python
summer_sports= merge_df[merge_df.Season == "Summer"].Sport.unique()
summer_sports
```

```python
gender_counts= merge_df.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize=(15,10))
plt.title("Gender Distribution")
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.2f%%", startangle=180, shadow=True);
```

```python
merge_df.Medal.value_counts()
```

```python
female_participants= merge_df[(merge_df.Sex=='F') & (merge_df.Season=='Summer')][['Sex', 'Year']]
female_participants= female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = merge_df[(merge_df.Sex == "F") & (merge_df.Season == "Summer")]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(15,10))
sns.countplot(x="Year", data=womenOlympics, palette="pastel")
plt.title("Women Participation");
```

```python
part = womenOlympics.groupby("Year")["Sex"].value_counts()
plt.figure(figsize=(15,10))
part.loc[:,"F"].plot()
plt.title("Plot of Female Athletes over time");
```

```python
goldmedals= merge_df[(merge_df.Medal== "Gold")]
goldmedals.head()
```

```python
goldmedals = goldmedals[np.isfinite(goldmedals["Age"])]
```

```python
goldmedals["ID"][goldmedals["Age"] > 60].count()
```

```python
sport_event= goldmedals["Sport"][goldmedals["Age"]>60]
sport_event
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.countplot(sport_event)
plt.title("Gold Medals for Athletes over 60 Years");
```

```python
goldmedals.Region.value_counts().reset_index(name="Medal").head(5)
```

```python
plt.figure(figsize=(15,10))
totalGoldMedals= goldmedals.Region.value_counts().reset_index(name="Medal").head(20)
g= sns.catplot(x="index", y="Medal", data=totalGoldMedals,
              height=7, kind="bar", palette="flare")
g.despine(left=True)
plt.xticks(rotation=60)
g.set_xlabels("Top 20 countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country");
```

Rio Olympics

```python
max_year = merge_df.Year.max()
print(max_year)

team_names= merge_df[(merge_df.Year == max_year) & (merge_df.Medal == "Gold")].Team

team_names.value_counts().head(10)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel("Country Medals for the Rio Olympics");
```
